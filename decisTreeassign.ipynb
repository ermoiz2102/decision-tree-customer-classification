# 🌳 Decision Tree Classification  

## Objective:
To build a Decision Tree model to predict customer behavior (defaulter or not) and understand decision rules using tree visualization.

## 📦 Importing Libraries  

Importing required libraries for data handling, model building, and evaluation.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

## 📂 Data Loading  

Loading dataset containing customer financial details.

## 📦 Importing Libraries  

Importing required libraries for data handling, model building, and evaluation.


```python
df=pd.read_csv("Decision_Tree_ Dataset.csv")
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

## 🔀 Train-Test Split  

Splitting dataset into training and testing sets to evaluate model performance.


## 🤖 Model Training  

Training Decision Tree Classifier with controlled depth to avoid overfitting.

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
model=DecisionTreeClassifier(max_depth=5)
model.fit(X_train,y_train)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
%matplotlib inline
```

## 🌳 Decision Tree Visualization  

Visualizing the tree to understand decision rules and feature importance.

```python
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
```

### 🔍 Observation:
The model splits data based on key features, creating decision rules that are easy to interpret.

## 📊 Model Prediction  

Predicting outcomes on test data.

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
```

## 📈 Model Evaluation  

Evaluating model using accuracy score and classification report.

```python
score
```

```python
print(classification_report(y_pred,y_test))
```

## 📊 Model Performance  

The Decision Tree model achieved an accuracy score of **0.9394 (93.94%)** on the test dataset after tuning the `max_depth` parameter to 5.

### 🔍 Observation:
- Limiting the tree depth helped reduce overfitting  
- The model generalizes well on unseen data  
- Depth = 5 provided the best balance between bias and variance
